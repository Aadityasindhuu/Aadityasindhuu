```python
import re
import nltk
import PyPDF2
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Load PDF
def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
    return text

# Preprocessing function
def preprocess_text(text):
    # Tokenize into sentences
    sentences = sent_tokenize(text)

    # Normalize: convert to lowercase and remove special characters
    normalized_sentences = [re.sub(r'[^a-zA-Z0-9\s]', '', sentence.lower()) for sentence in sentences]

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_sentences = [' '.join([word for word in sentence.split() if word not in stop_words])
                          for sentence in normalized_sentences]

    return filtered_sentences

# Example Usage:
# Upload a PDF file in Google Colab
from google.colab import files
uploaded = files.upload()

# Process the PDF
file_path = next(iter(uploaded.keys()))
text = read_pdf(file_path)
preprocessed_text = preprocess_text(text)

print(preprocessed_text[:10])  # Display first 10 preprocessed sentences
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments
from datasets import Dataset
import json

# Load GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Add padding token to the tokenizer
tokenizer.pad_token = tokenizer.eos_token  # or tokenizer.add_special_tokens({'pad_token': '[PAD]'})

# Load custom dataset (replace 'harry_potter_qa.json' with your actual file)
with open('harry_potter_qa.json', 'r') as f:
    harry_potter_data = json.load(f)

# Convert the data into a HuggingFace Dataset
def prepare_harry_potter_dataset(harry_potter_data):
    contexts = []
    questions = []
    answers = []

    for book in harry_potter_data['data']:
        for paragraph in book['paragraphs']:
            context = paragraph['context']
            for qa in paragraph['qas']:
                question = qa['question']
                answer = qa['answers'][0]['text']
                contexts.append(context)
                questions.append(question)
                answers.append(answer)

    return Dataset.from_dict({'context': contexts, 'question': questions, 'answer': answers})

# Prepare dataset
custom_dataset = prepare_harry_potter_dataset(harry_potter_data)

# Tokenization function (set consistent max length for input and output)
def preprocess_function(examples):
    inputs = ["question: " + q + " context: " + c for q, c in zip(examples['question'], examples['context'])]
    labels = examples['answer']

    # Tokenize inputs and labels with the same max length
    model_inputs = tokenizer(inputs, max_length=128, truncation=True, padding='max_length')

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(labels, max_length=128, truncation=True, padding='max_length')  # Set labels to the same max length

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Tokenize the dataset
tokenized_dataset = custom_dataset.map(preprocess_function, batched=True)

# Fine-tuning arguments
training_args = TrainingArguments(
    output_dir='./results',
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=2,  # You can adjust this based on available GPU memory
    save_steps=10_000,
    save_total_limit=2,
    logging_steps=500,
    report_to="none"
)

# Trainer object
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
)

# Fine-tune the model
trainer.train()

# Save the fine-tuned model
model.save_pretrained('./fine_tuned_gpt2_harry_potter')
tokenizer.save_pretrained('./fine_tuned_gpt2_harry_potter')
```

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

# Load the fine-tuned model and tokenizer
model = GPT2LMHeadModel.from_pretrained('./fine_tuned_gpt2_harry_potter')
tokenizer = GPT2Tokenizer.from_pretrained('./fine_tuned_gpt2_harry_potter')

# Load the FAISS index
index = faiss.read_index("faiss_vector_store.index")

# Load the original text chunks
with open("original_texts.txt", "r") as f:
    original_texts = f.readlines()

# Load the embedding model for query transformation
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

def retrieve_context(query, faiss_index, original_texts, k=5):
    query_embedding = embedding_model.encode([query])[0]
    D, I = faiss_index.search(np.array([query_embedding]), k)
    relevant_texts = [original_texts[i] for i in I[0]]

    # Print the retrieved context to verify its correctness
    print("Retrieved context:")
    print("\n".join(relevant_texts))

    # Concatenate relevant texts and truncate if needed
    context = " ".join(relevant_texts)
    return context[:1024]  # Truncate context to 1024 tokens or fewer

# Function to generate an answer using fine-tuned GPT-2 model
def generate_answer(query, context):
    input_text = f"Based on the following context, answer the question.\n\nContext: {context}\n\nQuestion: {query}\nAnswer:"

    # Tokenize the input with attention masks
    input_ids = tokenizer.encode(input_text, return_tensors="pt")
    attention_mask = torch.ones(input_ids.shape, dtype=torch.long)  # Create attention mask

    # Generate output
    output = model.generate(
        input_ids,
        attention_mask=attention_mask,
        max_length=150,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

    # Decode the generated answer
    answer = tokenizer.decode(output[0], skip_special_tokens=True)
    return answer

user_query = "When did Harry receive his letter from Hogwarts?"
context = retrieve_context(user_query, index, original_texts)
print("Context:", context)  # Verify context is relevant
answer = generate_answer(user_query, context)
print("Answer:", answer)
```

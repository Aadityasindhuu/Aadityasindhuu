```python
# The mode that i am chossing to go with is:
model = SentenceTransformer('intfloat/e5-large-v2')
```

```python
# Step 2: Create combined text and generate embeddings
def combine_text(row):
    return f"{row['bio']} Specialties: {row['specialties']}"

df['combined'] = df.apply(combine_text, axis=1)
df['embedding'] = df['combined'].apply(lambda x: model.encode(x, convert_to_tensor=True))
```

```python
# Step 3: Define user query function
def recommend_astrologers(user_input, top_k=3):
    user_embedding = model.encode(user_input, convert_to_tensor=True)

    df['score'] = df['embedding'].apply(lambda emb: float(cosine_similarity([user_embedding.cpu().numpy()], [emb.cpu().numpy()])[0][0]))
    top_matches = df.sort_values(by='score', ascending=False).head(top_k)

    results = []
    for _, row in top_matches.iterrows():
        results.append({
            "Astrologer": row['name'],
            "Specialties": row['specialties'],
            "Why Recommended": row['bio'],
            "Relevance Score": round(row['score'], 3)
        })
    return results
```

```python
# Step 4: Try an example
user_query = "I'm going through a breakup and want to know if my soulmate is out there or when I will find love again."

recommendations = recommend_astrologers(user_query)
for r in recommendations:
    print(f"\n🔮 Astrologer: {r['Astrologer']}")
    print(f"Specialties: {r['Specialties']}")
    print(f"Reason: {r['Why Recommended']}")
    print(f"Relevance Score: {r['Relevance Score']}")
```
